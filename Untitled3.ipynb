```python
age = int(input("Enter your age: "))

if age > 50:
    print("You are too old.")
elif age <= 50 and age >= 30:
    print("You are an adult.")
elif age <= 29 and age >= 18:
    print("You are young.")
elif age <= 17 and age >= 13:
    print("You are a teenager.")
elif age <= 12 and age >= 1:
    print("You are a kid.")
else:
    print("invalid")
```
